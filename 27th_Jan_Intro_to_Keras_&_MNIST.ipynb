```python
digit_avg = {}
for digit in range(10):
  digit_avg[digit] = np.mean(x_train[(y_train == digit)])
```

```python
digit_avg
```

```python
x_train_percentile = {}
for i in range(75, 101):
  x_train_percentile[i] = np.percentile(x_train, i)
plt.plot(x_train_percentile.keys(),x_train_percentile.values())
plt.xlabel("Percentile")
plt.ylabel("Values")
plt.title("X-Train Percentile x Values")
plt.show()
```

```python
for digit in range(10):
  percentile = []
  for i in range(70, 101):
    percentile.append(np.percentile(x_train[y_train == digit], i))
  plt.plot(np.arange(70, 101), percentile, label=digit)
  plt.legend()
  # plt.xticks(np.arange(75, 101))
  plt.xlabel("75%tile and ononwards")
  plt.title("Percentile of Digits")
```

```python
digit_zero = {}
for digit in range(10):
  digit_zero[digit] = (((x_train[(y_train == digit)]) == 0).sum())  / ((x_train[(y_train == digit)]).size)
digit_zero
```

```python
digit_30 = {}
for digit in range(10):
  digit_30[digit] = (((x_train[(y_train == digit)]) < 30).sum())  / ((x_train[(y_train == digit)]).size)
digit_30
```
